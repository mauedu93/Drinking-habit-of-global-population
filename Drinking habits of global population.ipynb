<b> Question 6:</b> Fit a linear regression model to predict the <code>'total_litres_of_pure_alcohol'</code> using the number of <code>'wine_servings'</code> then calculate $R^{2}$:


```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression() #create a linear regression object
#Define dependent variable, total litres of pure alcohol, and independent variable, wine servings.
x = df[['wine_servings']]
y = df['total_litres_of_pure_alcohol']

lm.fit(x,y) #fitting the regression
Yhat = lm.predict(x) #prediction of total litres of pure alcohol

print("The first five numbers of the prediction are:",Yhat[0:5])

print("The R-squared is:",lm.score(x,y)) 
```

### Question 7
Use the list of features to predict the <code>'total_litres_of_pure_alcohol'</code>, split the data into training and testing and determine the $R^2$ on the test data, using the provided code:

```python
from sklearn.linear_model import LinearRegression

#creating a dummy variable for continent
dummy_variable_1 = pd.get_dummies(df['continent'])
df = pd.concat([df,dummy_variable_1],axis=1) #insert dummy variable to original dataset
df.drop('continent',axis=1,inplace=True) #drop continent from the original dataset

#Linear model
y_data = df['total_litres_of_pure_alcohol'] #target data moved to a separate dataframe
x_data = df[['beer_servings','spirit_servings','wine_servings','Africa','Asia','Europe','North America','Oceania','South America']]

lm.fit(x_data,y_data) #fitting the linear model

#Making a prediction of Total litres of pure alcohol
yhat = lm.predict(x_data)
print("The first five elements of the prediction are:",yhat[0:5])

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state = 0) #spliting data into train and test

print("number of test samples:",x_test.shape[0])
print("number of training samples:",x_train.shape[0])

print("The R-squared on the test data is:",lm.score(x_test,y_test))
```

<b>Question 8 :</b> Create a pipeline object that scales the data, performs a polynomial transform and fits a linear regression model. Fit the object using the training data in the question above, then calculate the R^2 using. the test data. Take a screenshot of your code and the $R^{2}$. There are some hints in the notebook:

<code>'scale'</code>

<code>'polynomial'</code>

<code>'model'</code>

The second element in the tuple contains the model constructor

<code>StandardScaler()</code>

<code>PolynomialFeatures(include_bias=False)</code>

<code>LinearRegression()</code>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression

#pipeline object that scales the data
input = [('scale',StandardScaler()),('polynomial',PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
pipe = Pipeline(input)

pipe.fit(x_train,y_train) #fitting the model

print('The R-square is:',pipe.score(x_test,y_test))
```

<b>Question 9</b>: Create and fit a Ridge regression object using the training data, setting the regularization parameter to 0.1 and calculate the $R^{2}$ using the test data. Take a screenshot of your code and the $R^{2}$

```python
from sklearn.linear_model import Ridge

RigeModel = Ridge(alpha=0.1) #creating a Ridge Model
RigeModel.fit(x_train,y_train) #Fitting the Ridge regression

print("The R-squared of the model is:",RigeModel.score(x_test,y_test))
```

<b>Question 10 </b>: Perform a 2nd order polynomial transform on both the training data and testing data.  Create and fit a Ridge regression object using the training data, setting the regularization parameter to 0.1. Calculate the $R^{2}$ utilizing the test data provided. Take a screen-shot of your code and the $R^{2}$. 

```python
#Polynomial transformation of 2nd order for training and testing data
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train[['beer_servings','spirit_servings','wine_servings','Africa','Asia','Europe','North America','Oceania','South America']])
x_test_pr = pr.fit_transform(x_test[['beer_servings','spirit_servings','wine_servings','Africa','Asia','Europe','North America','Oceania','South America']])

#Ridge model of regularization parameter = 0.1
Rige_Model = Ridge(alpha=0.1)
Rige_Model.fit(x_train_pr,y_train) #Fitting the model

print("The R-squared of the test data is:",Rige_Model.score(x_test_pr,y_test))

```
